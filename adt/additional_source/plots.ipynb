```python
import os
import pandas as pd

# List of dictionaries (convert to DataFrames if necessary)
dfs = {
    '150_clean_time_15_counts': pd.DataFrame(df_d_150_time_15_counts),
    '150_clean_time_15_values': pd.DataFrame(df_d_150_time_15_values),
}

# Ensure output directory exists
output_dir = "parameters_best"
os.makedirs(output_dir, exist_ok=True)

# Metrics and parameters
metrics = [
    'precision', 
    'recall', 
    'llm_interaction_accuracy',
    'llm_activation_sensitivity', 
    'Tp',
    'Fp_in',
    'Fp_out',
    'Fn',
    'model_overall_accuracy'
]
falses = ['Fp', 'Fp_out', 'Fp_in', 'Fn']
parameter_columns = ['time', 'dot_value', 'distance_value', 'dot_counts', 'distance_counts']

# Iterate over each DataFrame
for sequence_name, df_d in dfs.items():
    # Ensure it's a DataFrame
    if not isinstance(df_d, pd.DataFrame):
        df_d = pd.DataFrame(df_d)
    
    # Max values
    max_precision = df_d[metrics[0]].max()  
    max_recall = df_d[metrics[1]].max()
    max_accuracy = df_d[metrics[2]].max()
    max_sensitivity = df_d[metrics[3]].max()
    max_overall = df_d[metrics[8]].max()

    # Filter rows
    best_rows_1 = df_d[(df_d[metrics[0]] == max_precision) & (df_d[metrics[2]] == max_accuracy)]
    best_rows_2 = df_d[(df_d[metrics[1]] == max_recall)]
    best_rows_3 = df_d[(df_d[metrics[3]] == max_sensitivity)]
    best_rows_4 = df_d[(df_d[metrics[2]] == max_accuracy)]
    best_rows_5 = df_d[(df_d[metrics[0]] == max_precision)]
    best_rows_6 = df_d[(df_d[metrics[8]] == max_overall)]

    # Save best parameter combinations
    best_combinations_precision_accuracy = best_rows_1[parameter_columns + [metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)
    best_combinations_recall = best_rows_2[parameter_columns + [metrics[0],metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)
    best_combinations_sensitivity = best_rows_3[parameter_columns + [metrics[0], metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)
    best_combinations_accuracy = best_rows_4[parameter_columns + [metrics[0], metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)
    best_combinations_precision = best_rows_5[ parameter_columns + [metrics[0], metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)
    best_combinations_overall = best_rows_6[parameter_columns + [metrics[0], metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]].reset_index(drop=True)

    # Save to CSV
    best_combinations_precision_accuracy.to_csv(f'{output_dir}/{sequence_name}_best_precision_accuracy.csv', index=False)
    best_combinations_recall.to_csv(f'{output_dir}/{sequence_name}_best_recall.csv', index=False)
    best_combinations_sensitivity.to_csv(f'{output_dir}/{sequence_name}_best_sensitivity.csv', index=False)
    best_combinations_accuracy.to_csv(f'{output_dir}/{sequence_name}_best_accuracy.csv', index=False)
    best_combinations_precision.to_csv(f'{output_dir}/{sequence_name}_best_precision.csv', index=False)
    best_combinations_overall.to_csv(f'{output_dir}/{sequence_name}_best_overall.csv', index=False)

    # Find minimum values for falses
    for false in falses:
        min_value = df_d[false].min()
        best_rows = df_d[df_d[false] == min_value]
        best_combination_false = best_rows[parameter_columns + [metrics[1], metrics[2], metrics[3], metrics[4], metrics[5], metrics[6], metrics[7], metrics[8]]+[false]].reset_index(drop=True)
        best_combination_false.to_csv(f'{output_dir}/{sequence_name}_best_{false}.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

# List of DataFrames (convert to DataFrames if necessary)
dfs = {
    '150_clean_time_15_counts': pd.DataFrame(df_d_150_time_15_counts),
    '150_clean_time_15_values': pd.DataFrame(df_d_150_time_15_values),
}

# Output directory for plots
output_dir = "plots_new"
os.makedirs(output_dir, exist_ok=True)

# List of metrics to plot
metrics_list = [
    'model_overall_accuracy',
    'precision',
    'recall',
    'llm_activation_sensitivity',
    'llm_interaction_accuracy',
    'Tp',
    'Fp_out',
    'Fp_in',
    'Fn'
]

# Define integer metrics
integer_metrics = ['Tp', 'Fp_out', 'Fp_in', 'Fp', 'Fn']

# Number of metrics
num_metrics = len(metrics_list)

# Determine the layout of subplots (adjust columns as needed)
num_cols = 3
num_rows = (num_metrics + num_cols - 1) // num_cols  # Ceiling division to get number of rows

# Loop through each DataFrame
for sequence_name, df_d in dfs.items():
    # Create subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(5 * num_cols, 4 * num_rows))
    axes = axes.flatten()  # Flatten to 1D array for easy indexing

    for idx, metric in enumerate(metrics_list):
        # Create pivot table for the current metric
        pivot_table = df_d.pivot_table(
            index='dot_value',
            columns='distance_value',
            values=metric,
            aggfunc='mean'  # Using mean results in float values
        )
        # Sort the index and columns for consistent plotting
        pivot_table = pivot_table.sort_index().sort_index(axis=1)

        # Determine formatting and scaling based on metric type
        if metric in integer_metrics:
            fmt = ".2f"          # Float format with zero decimal places
            vmin, vmax = None, None  # Auto-scale
            cmap = 'OrRd'        # Colormap suitable for counts
        else:
            fmt = ".2f"          # Float format with two decimal places
            vmin, vmax = 0, 1    # Fixed scale for float metrics between 0 and 1
            cmap = 'YlGnBu'      # Colormap suitable for percentages

        # Plot heatmap
        sns.heatmap(
            pivot_table,
            ax=axes[idx],
            annot=True,
            fmt=fmt,
            cmap=cmap,
            cbar=False,
            vmin=vmin,
            vmax=vmax,
            annot_kws={"size": 12, "weight": "bold"}  # Larger and bold text for annotations
        )
        axes[idx].set_title(metric.replace('_', ' ').title(), fontsize=14, fontweight='bold')  # Larger title
        axes[idx].set_xlabel('Distance Value', fontsize=12, fontweight='bold')  # Larger x-axis label
        axes[idx].set_ylabel('Dot Value', fontsize=12, fontweight='bold')  # Larger y-axis label
        axes[idx].tick_params(axis='both', which='major', labelsize=10)  # Larger tick labels
        # Rotate x-axis labels for better readability
        axes[idx].set_xticklabels(axes[idx].get_xticklabels(), rotation=45, ha='right', fontsize=10, fontweight='bold')
        axes[idx].set_yticklabels(axes[idx].get_yticklabels(), fontsize=10, fontweight='bold')

    # Remove any unused subplots
    for idx in range(num_metrics, len(axes)):
        fig.delaxes(axes[idx])

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Save the figure with sequence name in the file name
    output_path = f"{output_dir}/{sequence_name}_metrics_heatmap.png"
    plt.savefig(output_path, dpi=300, bbox_inches='tight')

    # Show the plot
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

# List of dictionaries (convert to DataFrames if necessary)
dfs = {
    '150_clean_time_15_counts': pd.DataFrame(df_d_150_time_15_counts),
    '150_clean_time_15_values': pd.DataFrame(df_d_150_time_15_values),
}

# Output directory for plots
output_dir = "plots_new"
os.makedirs(output_dir, exist_ok=True)

# Metrics to plot
line_metrics = [
    'precision',
    'recall',
    'llm_activation_sensitivity',
    'llm_interaction_accuracy',
    'model_overall_accuracy'
]

# Loop over each DataFrame
for sequence_name, df_d in dfs.items():
    # Melt the DataFrame for seaborn, including both 'dot_value' and 'distance_value' in id_vars
    metrics = df_d.melt(
        id_vars=['dot_value', 'distance_value'],
        value_vars=line_metrics,
        var_name='Metric',
        value_name='Value'
    )

    # Ensure 'dot_value' and 'distance_value' are numeric
    metrics['dot_value'] = pd.to_numeric(metrics['dot_value'], errors='coerce')
    metrics['distance_value'] = pd.to_numeric(metrics['distance_value'], errors='coerce')

    # Drop rows with NaN in 'dot_value' or 'distance_value' if any
    metrics.dropna(subset=['dot_value', 'distance_value'], inplace=True)

    # Sort the data by Metric and the respective x-axis for consistent plotting
    metrics.sort_values(['Metric', 'dot_value'], inplace=True)

    # Create the figure and axes for subplots
    fig, ax = plt.subplots(2, 1, figsize=(14, 15), sharex=False)  # sharex=False since different x-axes

    # ----------------------------
    # First Subplot: Line Metrics vs. Dot Value
    # ----------------------------
    sns.lineplot(
        data=metrics,
        x='dot_value',
        y='Value',
        hue='Metric',
        marker='o',
        ax=ax[0],
        linewidth=2.5,
        markersize=8
    )

    # Set bold font for titles and labels with larger font size
    ax[0].set_ylabel('Value (Line Metrics)', fontweight='bold', fontsize=14)
    ax[0].set_xlabel('Dot Value', fontweight='bold', fontsize=14)
    ax[0].set_title(f'Performance Metrics vs. Dot Value (Time 1.5sec)', fontweight='bold', fontsize=16)
    ax[0].tick_params(axis='both', which='major', labelsize=12)

    # ----------------------------
    # Second Subplot: Line Metrics vs. Distance Value
    # ----------------------------
    sns.lineplot(
        data=metrics,
        x='distance_value',
        y='Value',
        hue='Metric',
        marker='o',
        ax=ax[1],
        linewidth=2.5,
        markersize=8
    )

    # Set bold font for titles and labels with larger font size
    ax[1].set_ylabel('Value (Line Metrics)', fontweight='bold', fontsize=14)
    ax[1].set_xlabel('Distance Value', fontweight='bold', fontsize=14)
    ax[1].set_title(f'Performance Metrics vs. Distance Value (Time 1.5sec)', fontweight='bold', fontsize=16)
    ax[1].tick_params(axis='both', which='major', labelsize=12)

    # ----------------------------
    # Legend Customization
    # ----------------------------
    handles, labels = ax[0].get_legend_handles_labels()
    ax[0].get_legend().remove()
    ax[1].get_legend().remove()

    # Create a single legend for both plots
    fig.legend(
        handles, labels,
        title='Performance Metrics',
        fontsize=12,
        title_fontsize=14,
        loc='upper center',
        ncol=5
    )

    # Adjust layout to make room for the legend
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust the rect to leave space for the legend at the top

    # Save the figure with sequence name in the file name
    output_path = f"{output_dir}/{sequence_name}_metrics_dot_distance_value.png"
    plt.savefig(output_path, dpi=300, bbox_inches='tight')

    # Show the plot
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

# Metrics to plot
line_metrics = [
    'precision',
    'recall',
    'llm_activation_sensitivity',
    'llm_interaction_accuracy',
    'model_overall_accuracy'
]

# Melt the DataFrame for seaborn, including 'minimum_time', 'maximum_time', and 'movement' in id_vars
metrics = df_d_150_ra.melt(
    id_vars=['minimum_time', 'maximum_time', 'movement'],
    value_vars=line_metrics,
    var_name='Metric',
    value_name='Value'
)

# Output directory for plots
output_dir = "plots_new"
sequence_name = "150_clean"
os.makedirs(output_dir, exist_ok=True)

# Ensure 'minimum_time', 'maximum_time', and 'movement' are numeric
metrics['minimum_time'] = pd.to_numeric(metrics['minimum_time'], errors='coerce')
metrics['maximum_time'] = pd.to_numeric(metrics['maximum_time'], errors='coerce')
metrics['movement'] = pd.to_numeric(metrics['movement'], errors='coerce')

# Drop rows with NaN in 'minimum_time', 'maximum_time', or 'movement'
metrics.dropna(subset=['minimum_time', 'maximum_time', 'movement'], inplace=True)

# Sort the data by Metric and the respective x-axis for consistent plotting
metrics.sort_values(['Metric', 'minimum_time'], inplace=True)

# Create the figure and axes for subplots
fig, ax = plt.subplots(3, 1, figsize=(14, 18), sharex=False)  # Adjusted height for 3 subplots

# ----------------------------
# First Subplot: Line Metrics vs. Minimum Time
# ----------------------------
sns.lineplot(
    data=metrics,
    x='minimum_time',
    y='Value',
    hue='Metric',
    marker='o',
    ax=ax[0],
    linewidth=2.5,
    markersize=8
)

# Set bold font for titles and labels with larger font size
ax[0].set_ylabel('Value (Line Metrics)', fontweight='bold', fontsize=14)
ax[0].set_xlabel('Minimum Time', fontweight='bold', fontsize=14)
ax[0].set_title('Line Metrics vs. Minimum Time', fontweight='bold', fontsize=16)
ax[0].tick_params(axis='both', which='major', labelsize=12)

# ----------------------------
# Second Subplot: Line Metrics vs. Maximum Time
# ----------------------------
sns.lineplot(
    data=metrics,
    x='maximum_time',
    y='Value',
    hue='Metric',
    marker='o',
    ax=ax[1],
    linewidth=2.5,
    markersize=8
)

# Set bold font for titles and labels with larger font size
ax[1].set_ylabel('Value (Line Metrics)', fontweight='bold', fontsize=14)
ax[1].set_xlabel('Maximum Time', fontweight='bold', fontsize=14)
ax[1].set_title('Line Metrics vs. Maximum Time', fontweight='bold', fontsize=16)
ax[1].tick_params(axis='both', which='major', labelsize=12)

# ----------------------------
# Third Subplot: Line Metrics vs. Movement
# ----------------------------
sns.lineplot(
    data=metrics,
    x='movement',
    y='Value',
    hue='Metric',
    marker='o',
    ax=ax[2],
    linewidth=2.5,
    markersize=8
)

# Set bold font for titles and labels with larger font size
ax[2].set_ylabel('Value (Line Metrics)', fontweight='bold', fontsize=14)
ax[2].set_xlabel('User Movement', fontweight='bold', fontsize=14)
ax[2].set_title('Line Metrics vs. Movement', fontweight='bold', fontsize=16)
ax[2].tick_params(axis='both', which='major', labelsize=12)

# ----------------------------
# Legend Customization
# ----------------------------
# Get handles and labels from the first subplot
handles, labels = ax[0].get_legend_handles_labels()

# Remove legends from individual subplots
ax[0].get_legend().remove()
ax[1].get_legend().remove()
ax[2].get_legend().remove()

# # Create a single legend below all subplots
# fig.legend(handles, labels, title='Line Metrics', fontsize=12, title_fontsize=14, loc='upper center', ncol=5)

# # Adjust layout to make room for the legend
# plt.tight_layout(rect=[0, 0, 1, 0.93])  # Adjust the rect to leave space for the legend

# Create a single legend below the figure title and above the first plot
fig.legend(handles, labels, title='Line Metrics', fontsize=12, title_fontsize=14,
           loc='upper center', bbox_to_anchor=(0.5, 0.95), ncol=5)

# Adjust layout to make room for the legend and the title
plt.tight_layout(rect=[0, 0, 1, 0.9])  # Adjust the rect to leave space for the title and legend

# Add the sequence name to the figure title
fig.suptitle(f'Performance Metrics for {sequence_name}', fontweight='bold', fontsize=18, y=0.97)

# Save the figure with sequence name in the file name
output_path = f"{output_dir}/{sequence_name}_reactivation_metrics.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
